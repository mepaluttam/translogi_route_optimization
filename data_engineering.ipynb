```python
import pandas as pd
```

```python
df = pd.read_csv("start_data.csv")
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df['pickup_time'] = pd.to_datetime(df['pickup_time'])
df['actual_delivery'] = pd.to_datetime(df['actual_delivery'])
df['estimated_delivery'] = pd.to_datetime(df['estimated_delivery'])
df['earliest_delivery'] = pd.to_datetime(df['earliest_delivery'])
df['latest_delivery'] = pd.to_datetime(df['latest_delivery'])
# Calculate actual delivery time in hours
df['actual_delivery_time_hours'] = (df['actual_delivery'] - df['pickup_time']).dt.total_seconds() / 3600

# Reorder columns to place the new column right after 'actual_delivery'
cols = df.columns.tolist()
cols.remove('actual_delivery_time_hours')  # Ensure it doesn't exist twice
new_order = cols[:cols.index('actual_delivery') + 1] + ['actual_delivery_time_hours'] + cols[cols.index('actual_delivery') + 1:]
df = df[new_order]
```

```python
df
```

```python
def remove_outliers(df):
    # Create an empty list to store columns with outliers removed
    df_cleaned = df.copy()
    
    # Loop over numerical columns to remove outliers
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        # Calculate the first and third quartiles
        Q1 = df_cleaned[column].quantile(0.25)
        Q3 = df_cleaned[column].quantile(0.75)
        
        # Calculate the Interquartile Range (IQR)
        IQR = Q3 - Q1
        
        # Define the upper and lower bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Remove the rows where the column values are outliers
        df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]
    
    return df_cleaned

df_cleaned = remove_outliers(df)  # Remove outliers from the dataframe

```

Normalize Geolocation Data 

```python
from sklearn.preprocessing import MinMaxScaler

# Assume `df` is your DataFrame
scaler = MinMaxScaler()

# Normalize latitude and longitude columns
df_cleaned[['pickup_lat', 'pickup_lon', 'delivery_lat', 'delivery_lon']] = scaler.fit_transform(
    df_cleaned[['pickup_lat', 'pickup_lon', 'delivery_lat', 'delivery_lon']]
)
```

Average Delivery Time by Area (City)

```python
# Calculate the delivery time difference
# Convert 'estimated_delivery' to datetime if it's a string
df_cleaned['delivery_time_diff'] = (df_cleaned['actual_delivery'] - df_cleaned['estimated_delivery']).dt.total_seconds() / 3600


# Calculate average delivery time by 'delivery_city' 
avg_delivery_time_by_city = df_cleaned.groupby('delivery_city')['delivery_time_diff'].mean()
```

```python
# Correlation between traffic level and delivery time
traffic_impact = df_cleaned[['traffic_level_category', 'delivery_time_diff']].groupby('traffic_level_category').mean()

# If you want to include weather as well:
weather_impact = df_cleaned[['weather_main', 'delivery_time_diff']].groupby('weather_main').mean()
```

```python
# Calculate capacity utilization
df_cleaned['weight_utilization(%)'] = df_cleaned['weight_kg'] / df_cleaned['vehicle_capacity_weight'] * 100
df_cleaned['volume_utilization(%)'] = df_cleaned['volume_m3'] / df_cleaned['vehicle_capacity_volume'] * 100
```
